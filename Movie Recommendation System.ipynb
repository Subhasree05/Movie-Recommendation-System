```python
#importing dependencies
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#data collection and preprocessing
movies=pd.read_csv('C:/users/biswa/Downloads/movies.csv')
```

```python
movies.head()
```

```python
movies.shape
```

```python
#selecting relevant features
selected=['genres','keywords','tagline','cast','director']
print(selected)
```

```python
#replacing null values w/ null string
for feature in selected:
    movies[feature]=movies[feature].fillna(' ')
```

```python
#combining selected features
combined=movies['genres']+' '+movies['keywords']+' '+movies['tagline']+' '+movies['cast']+' '+movies['director']
print(combined)
```

```python
vect=TfidfVectorizer()
```

```python
featurevect=vect.fit_transform(combined)
```

```python
print(featurevect)
```

```python
#cosine similarity
sim=cosine_similarity(featurevect)
```

```python
print(sim)
```

```python
sim.shape
```

```python
#getting movie name from user
moviename=input("Enter your favourite movie name : ")
```

```python
#create a list of movies
listoftitles=movies['title'].tolist()
print(listoftitles)
```

```python
#find close match for user input
findmatch=difflib.get_close_matches(moviename,listoftitles)
print(findmatch)
```

```python
closematch=findmatch[0]
print(closematch)
```

```python
#finding index of movie
indexmov=movies[movies.title==closematch]['index'].values[0]
print(indexmov)
```

```python
#getting list of similar movies
simscore=list(enumerate(sim[indexmov]))
print(simscore)
```

```python
len(simscore)
```

```python
#sort based on simscore
sortedmov=sorted(simscore,key=lambda x:x[1],reverse=True)
print(sortedmov)
```

```python
#print name of similar movies based on index
print("Movies suggested for you : \n")
i=1
for mov in sortedmov:
    index=mov[0]
    title_from_index=movies[movies.index==index]['title'].values[0]
    if(i<=30):
        print(i,'.',title_from_index)
        i+=1
```

```python
#Movie Recommendation System
moviename=input("Enter your favourite movie name : ")

listoftitles=movies['title'].tolist()

findmatch=difflib.get_close_matches(moviename,listoftitles)

closematch=findmatch[0]

indexmov=movies[movies.title==closematch]['index'].values[0]

simscore=list(enumerate(sim[indexmov]))

sortedmov=sorted(simscore,key=lambda x:x[1],reverse=True)

print("Movies suggested for you : \n")
i=1
for mov in sortedmov:
    index=mov[0]
    title_from_index=movies[movies.index==index]['title'].values[0]
    if(i<=30):
        print(i,'.',title_from_index)
        i+=1
```
